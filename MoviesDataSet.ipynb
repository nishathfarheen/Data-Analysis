```python
import pandas as pd
import numpy as np
mo=pd.read_csv("/content/movies.csv")
mo.head()
```

```python

mo.drop(mo[mo["Release Date"]== "TBD"].index,inplace=True)
from datetime import datetime
mo["Release Date"]=pd.to_datetime(mo["Release Date"])
```

```python
np.array(mo["Release Date"])
```

```python
#1.	 (i). Find out the no of movies released in every month of the year 1995.

new_df=mo[["Title","Release Date"]]
st_date=datetime.strptime("01-Jan-95","%d-%b-%y")
en_date=datetime.strptime("31-Dec-95","%d-%b-%y")
df1=new_df.loc[(new_df["Release Date"]>=st_date) & (new_df["Release Date"]<=en_date)]

df2=df1.groupby(df1["Release Date"].dt.month).count()
df2["Title"]
```

```python
#(ii). Find out the no. of movies released in every year from 1990 to 1998.
re1=mo.loc[(mo["Release Date"].dt.year>=1990) & (mo["Release Date"].dt.year<=1998)]
a1=re1.groupby(mo["Release Date"].dt.year).count()
a1["Title"]
```

```python
#2.	(i). Find no. of movies released under each genre given in the database.
df=mo.groupby("Major Genre").count()
df["Title"]
```

```python
#(ii). Find the movies under each genre with 1MDB rating >7 and rotten tomatoes rating > 60.
df1=mo.loc[(mo["IMDB Rating"]>7) & (mo["Rotten Tomatoes Rating"]>60)]
df2=df1.groupby("Major Genre")
df2["Title"].apply(list)
```

```python
#3.	(i). Find the movies released under each fiction with each director in the ascending order of release dates.
k=mo.sort_values(by="Release Date",ascending=True)
mo1=k.groupby(["Creative Type","Director","Release Date"])
mo1["Title"].apply(list)
```

```python
#(ii). Find movies released under each distributor in the order of genre and director.
p=mo.sort_values(by=["Major Genre","Director"],ascending=True)
mo2=p.groupby(["Distributor","Major Genre","Director"])
mo2["Title"].apply(list)
```

```python
#4.	(i). Find the movies released world-wide and find out the revenue received world-wide other than US with their ratings.
mo["US Gross"]=pd.to_numeric(mo["US Gross"],errors='coerce')
mo["Worldwide Gross"]=pd.to_numeric(mo["Worldwide Gross"],errors='coerce')
f1=mo[["Title","US Gross","Worldwide Gross","IMDB Rating"]]
f1["differ"]=mo["Worldwide Gross"]-mo["US Gross"]
f1
```

```python
#(ii). Find the movies with loss & profit released in each year with genre and ratings.
mo2=mo[["Title","IMDB Rating","Major Genre"]]
mo2["Loss and Profit"]=mo["Worldwide Gross"]-mo["Production Budget"]
mo2
```
