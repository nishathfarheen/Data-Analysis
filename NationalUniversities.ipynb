```python
import pandas as pd
nf=pd.read_csv("/content/National_Universities_Rankings.csv")
nf
```

```python
#1.	Find the universities in which undergraduate students were admitted.
print("The universities in which undergraduate students were admitted are:")
nf["Name"]
```

```python
#2.	List the states along with the cities in which universities located.
nf[["Name","Location"]]
```

```python
##3.	List the cities &universities under each state.
nf["city"]=nf["Location"].str.split(',').str[0]
nf["state"]=nf["Location"].str.split(',').str[1]
nf.head()
```

```python
cit=nf.groupby(["state"]).agg(lambda x: list(x))
cit[["city","Name"]]
```

```python
#4.	How many universities in each state have ranking<50? Print them along with ranking.
nf.loc[nf["Rank"]<50,["Name","Rank"]]
```

```python
from numpy import NaN
#5. How many universities have both out-of state and in-state students?
print("The universities that have both out-of state and in-state students",len(nf.loc[:].index)-nf.isna().sum()["In-state"])
```

```python
nf.loc[nf["In-state"].isna()==False]
```

```python
#6.	How many universities have marginal difference <=$5000 in in-state &out-of state tuition fees.
nf["Tuition and fees"]=nf["Tuition and fees"].str[1:]
nf["Tuition and fees"]=nf["Tuition and fees"].str.replace(",","",1)
nf
```

```python
nf["In-state"]=nf["In-state"].str[1:]
nf["In-state"]=nf["In-state"].str.replace(",","",1)
nf["In-state"]=nf["In-state"].fillna(0)
nf
```

```python
nf["Tuition and fees"]=nf["Tuition and fees"].astype(int)
nf["In-state"]=nf["In-state"].astype(int)
```

```python
nf.loc[(nf["Tuition and fees"]-nf["In-state"])<=5000,"Name"]
```

```python
#7. List the universities having tuition fee>15000$ and rank between 120 to 170
nf.loc[(nf["Tuition and fees"]>15000) & (nf["Rank"]>=120) & (nf["Rank"]<=170),"Name"]
```

```python
#8.	Find the campuses of universities located in different cities (multiple cities)
gk=nf.groupby("city")
gk["Name"].first()
```

```python
#9.	Mention the states where out-of state fee is more than in-state students. Print Minimum and Maximum fees.
print("the states where out-of state fee is more than in-state students",nf.loc[nf["Tuition and fees"]>nf["In-state"],"state"])
print("Maximum fees:",nf["Tuition and fees"].max())
print("Minimum fees:",nf["Tuition and fees"].min())
```

```python
#10.	Find the cities locating top 100 universities.
nf.loc[nf["Rank"]<=100,"city"]
```

```python
#11.	Find universities with least no of undergraduate students.
nf.loc[nf["Undergrad Enrollment"]==nf["Undergrad Enrollment"].min(), "Name"]
```

```python
#12.	Identifying correlations between enrollment numbers and university rank
nf["Undergrad Enrollment"]=nf["Undergrad Enrollment"].str.replace(',',"",1).astype(int)
```

```python
nf["Undergrad Enrollment"].corr(nf["Rank"])
```
